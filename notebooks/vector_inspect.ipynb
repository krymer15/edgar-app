```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
import psycopg2
from dotenv import load_dotenv

load_dotenv()

conn = psycopg2.connect(os.getenv("PGVECTOR_DB_URL"))

# Filing summaries
filing_df = pd.read_sql("SELECT * FROM filing_summaries ORDER BY filing_date DESC", conn)

# Exhibit summaries
exhibit_df = pd.read_sql("SELECT * FROM exhibit_summaries ORDER BY filing_date DESC", conn)

conn.close()

# üîç Display
print("üßæ Filing Summaries")
display(filing_df)

print("\nüìé Exhibit Summaries")
display(exhibit_df)
```
